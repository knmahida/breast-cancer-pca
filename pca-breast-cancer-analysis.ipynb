```python
# Import plotting and data libraries; enable inline plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import dataset loader for the breast cancer dataset
from sklearn.datasets import load_breast_cancer
```

```python
# Load the breast cancer dataset into the variable 'cancer'
cancer = load_breast_cancer()
```

```python
# List the keys available in the dataset object
cancer.keys()
```

```python
# Print the dataset description and metadata to understand features and classes
print(cancer['DESCR'])
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target_names']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scalered_data=scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scalered_data)
```

```python
x_pca=pca.transform(scalered_data)
```

```python
scalered_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Breast Cancer Dataset')
plt.colorbar()
plt.show()
```

```python
pca.components_
```

```python
df_components=pd.DataFrame(pca.components_,columns=cancer['feature_names'],index=['PC1','PC2'])
plt.figure(figsize=(12,6))
sns.heatmap(df_components,cmap='plasma',annot=True)
plt.title('PCA Component Heatmap')
plt.show()
```
